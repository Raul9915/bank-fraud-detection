```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/bank-fraud-data/Comprehensive_Banking_Database.csv'
df = pd.read_csv(path)
df.head()
```

```python
#Reemplazar Anomaly: -1 (fraude) -> 1, 1 (normal) -> 0
df['Anomaly'] = df['Anomaly'].apply(lambda x: 1 if x == -1 else 0)
```

```python
#Eliminar columnas irrelevantes
drop_cols = [
    'Customer ID', 'First Name', 'Last Name', 'Address', 'Email',
    'Contact Number', 'TransactionID', 'Loan ID', 'CardID', 'Feedback ID'
]
df.drop(columns=drop_cols, inplace=True)
```

```python
#Convertir columnas de fecha
date_cols = [
    'Date Of Account Opening', 'Last Transaction Date', 'Transaction Date',
    'Approval/Rejection Date', 'Payment Due Date',
    'Last Credit Card Payment Date', 'Feedback Date', 'Resolution Date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
#Verificar valores nulos
print(df.isnull().sum().sort_values(ascending=False))
```

```python
#Exportar una muestra para GitHub (opcional)
df.sample(100, random_state=42).to_csv('sample_Comprehensive_Banking_Database.csv', index=False)
```
